```python
import pandas as pd
from sqlalchemy import create_engine
import kaggle
import zipfile
import os
from dotenv import load_dotenv

# 1. Download the latest version of the dataset 
!kaggle datasets download -d vivek468/superstore-dataset-final

# 3. Extract the downloaded ZIP file
with zipfile.ZipFile('superstore-dataset-final.zip', 'r') as zip_ref:
    zip_ref.extractall('superstore_data')

# 4. Load the data into a Pandas DataFrame 
df = pd.read_csv('superstore_data/Sample - Superstore.csv', encoding='windows-1252')

#df.columns
##columns rename
df.columns = (df.columns.str.replace(" ", "").str.replace("-", ""))

#change the data types
df['OrderDate'] = pd.to_datetime(df['OrderDate'])
df['ShipDate'] = pd.to_datetime(df['ShipDate'])
df['Sales'] = df['Sales'].astype(float)
df['Profit'] = df['Profit'].astype(float)
df['Quantity'] = df['Quantity'].astype(int)
df['Discount'] = df['Discount'].astype(float)
df['PostalCode'] = df['PostalCode'].fillna("").astype(str).str.zfill(5)

#import to MS SQL Server
load_dotenv()
server = os.getenv('DB_SERVER')
database = os.getenv('DB_NAME')
driver = 'ODBC Driver 17 for SQL Server'

connection_string = f"mssql+pyodbc://{server}/{database}?driver={driver}&trusted_connection=yes"
engine = create_engine(connection_string)

df.to_sql("stg_Superstore", engine, if_exists="replace", index=False)
print("Data uploaded to Staging_Sales table in SSMS")
```
